This ipython notebook extensively focuses on importing and performing **Exploratory Data Analysis (EDA)** on the Californing House Pricing dataset from scikit-learn using basic pandas, matplotlib and seaborn.

```python
# Importing and loading dataset

from sklearn.datasets import fetch_california_housing
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

data = fetch_california_housing(as_frame=True)
df = data.frame
```

```python
# Target Variable Distribution

plt.figure(figsize=(6,4))
sns.histplot(df["MedHouseVal"], bins=50, kde=True)
plt.title("Distribution of Median House Value")
plt.show()
```

```python
# Features Distribution

df.hist(figsize=(14,10), bins=40)
plt.suptitle("Feature Distributions", fontsize = 16)
plt.tight_layout()
plt.show()
```

```python
# Correlation Analysis (Heatmap)

plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df)
plt.xticks(rotation=45)
plt.title("Outliers Before Handling")
plt.show()
```

```python
def iqr_clip(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[col] = df[col].clip(lower, upper)

outlier_cols = ["Population", "AveRooms", "AveBedrms", "AveOccup"]

for col in outlier_cols:
    iqr_clip(df, col)
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=df[outlier_cols])
plt.xticks(rotation=45)
plt.title("Outliers After IQR Clipping")
plt.show()
```

# üîç **Conclusion and Key Takeaways**

- **`MedInc` ‚Üí** strongest predictor (0.69 correlation coefficient with **`MedHouseVal`**)
- Latitude & Longitude matter.
- Rooms-related features and population have weak correlation.

Hence we will not do this outlier handling in our final program because the outliers (like `Population`, `AveOccup`) handled here have very weak or negligible correlation/affect on the **`MedHouseVal`**.
